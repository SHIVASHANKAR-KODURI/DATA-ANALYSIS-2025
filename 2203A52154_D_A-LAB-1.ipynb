```python
import pandas as pd
```

```python
a=pd.read_csv('/content/insurance.csv')
print(a)
```

a) Create vectors, factors, lists, matrices, and data frames.

```python
#vector
age_vector = a['age']
#factor
smoker_factor = pd.factorize(a['smoker'])[0]
#List
regions = a['region'].unique().tolist()
#matrix
age_bmi_matrix = a[['age', 'bmi']].values
#dataframe
filtered_df = a[a['age'] > 30]
```

b) Generate a scatter plot. Customize it by adding a title, axis labels, text annotations, and changing the plot characters (e.g., circles to plus signs).

```python
import matplotlib.pyplot as plt
plt.scatter(a['age'], a['expenses'], marker='+', color='red')
plt.title('Scatter Plot of Age vs. expenses')
plt.xlabel('Age')
plt.ylabel('expenses')
plt.show()
```

c) Filter the data based on a condition and updates the plot to reflect the filtered data.

```python
filtered_data = a[a['smoker'] == 'yes']

plt.scatter(filtered_data['age'], filtered_data['expenses'], marker='o', color='green')
plt.title('Scatter Plot of Age vs. expenses (Smokers)')
plt.xlabel('Age')
plt.ylabel('expenses')
plt.show()
```

```python
#Plot a scatter plot between age and smoking,
plt.scatter(a['age'], a['smoker'])
plt.title('Scatter Plot of Age vs. Smoking')
plt.xlabel('Age')
plt.ylabel('Smoker')
plt.show()
```

```python
#Plot a scatter plot between age(20 to 40) and smoking
filtered_age = a[(a['age'] >= 20) & (a['age'] <= 40)]
plt.scatter(filtered_age['age'], filtered_age['smoker'])
plt.title('Scatter Plot of Age (20-40) vs. Smoking')
plt.xlabel('Age')
plt.ylabel('Smoker')
plt.show()
```

```python
#Find expenses of smokers and non smokers,(plot)
smoker_expenses = a.groupby('smoker')['expenses'].mean()
plt.bar(smoker_expenses.index, smoker_expenses.values)
plt.title('Average Expenses by Smoking Status')
plt.xlabel('Smoking Status')
plt.ylabel('Average Expenses')
plt.show()
```

```python
#Find the number of smokers gender wise (plot)
smoker_gender_counts = a[a['smoker'] == 'yes']['sex'].value_counts()
plt.pie(smoker_gender_counts, labels=smoker_gender_counts.index, autopct='%1.1f%%')
plt.title('Distribution of Smokers by Gender')
plt.show()
```
